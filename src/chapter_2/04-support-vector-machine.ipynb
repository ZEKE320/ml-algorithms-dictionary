```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
```

```python
centers = np.array([(-1, -0.125), (0.5, 0.5)], dtype=np.float32)

X, y = make_blobs(n_samples=50, n_features=2, centers=centers, cluster_std=0.3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# MatplotlibでX_trainとX_testをプロット
# yが0の場合は青、1の場合は赤
import matplotlib.pyplot as plt

plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color="blue")
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color="red")
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color="lightblue")
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color="pink")
plt.show()
```

```python
model = LinearSVC()

y_pred = model.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
svc_slope: np.float_ = -model.coef_[0][0] / model.coef_[0][1]
svc_intercept: np.float_ = -model.intercept_[0] / model.coef_[0][1]

x = np.linspace(X[:, 0].min(), X[:, 0].max())
plt.ylim(X[:, 1].min() - 0.1, X[:, 1].max() + 0.1)
plt.plot(x, svc_slope * x + svc_intercept, color="green")
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color="blue")
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color="red")
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color="lightblue")
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color="pink")
plt.show()
```
