```python
import numpy as np
from numpy.typing import NDArray
from sklearn.linear_model import LogisticRegression

X_train = np.r_[
    np.random.normal(3, 1, size=50), np.random.normal(-1, 1, size=50)
].reshape((100, -1))

y_train = np.r_[np.ones(50), np.zeros(50)]
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict_proba(np.array([[0], [1], [2]]))[:, 1]
```

```python
import matplotlib.pyplot as plt

xx: NDArray = np.linspace(-4, 4, 1000)
proba: NDArray = model.predict_proba(xx.reshape(-1, 1))[:, 1]

plt.plot(xx, proba)
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Probability of Needing Winter Boots")
plt.show()
```
